```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

import os, sys
module_path = os.path.abspath(os.path.join('../..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from utils import  DBtable_to_df
```

```python
df = DBtable_to_df('../../nationalities.duckdb','nationalitiesClean')
```

## Outliers

```python
# Distribution plots
fig, axes = plt.subplots(6, 4, figsize=(15, 15))
for i, c in enumerate(df.columns[0:24]):
    ax = axes.reshape(-1)[i]
    sns.distplot(df[c],ax=ax)

    c = c.replace('_', ' ')
    title = f'{c[:35]}-\n{c[35:]}' if len(c) > 35 else c
    t = ax.set_title(title)
    ax.set(xlabel=None)  # remove the axis label
plt.tight_layout()
```

```python
# Boxplots
fig, axes = plt.subplots(6, 4, figsize=(15, 15))
for i, c in enumerate(df.columns[0:24]):
    ax = axes.reshape(-1)[i]
    sns.boxplot(df[c],ax=ax)

    c = c.replace('_', ' ')
    title = f'{c[:35]}-\n{c[35:]}' if len(c) > 35 else c
    t = ax.set_title(title)
plt.tight_layout()
```
